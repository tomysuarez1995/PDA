    ```python
    np.random.seed(42)
    a = np.random.randint(20, size=10)
    ## a es array([ 6, 19, 14, 10,  7,  6, 18, 10, 10,  3])
    
    ## salida esperada
    [1, 9, 7, 4, 3, 2, 8, 5, 6, 0]
    ```
    


````python
import numpy as np
def mediaMovil(val1, val2):
    ret = np.cumsum(val1, dtype=float)
    ret[val2:] = ret[val2:] - ret[:-val2]
    return ret[val2 - 1:] / val2

val1=np.arange(1,26)
val1=np.array(val1)
coso= mediaMovil(val1, val2=10)
print(coso)
````
