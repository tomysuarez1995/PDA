    ```python
    np.random.seed(42)
    ```

````python
import numpy as np
def mediaMovil(val1, val2):
    ret = np.cumsum(val1, dtype=float)
    ret[val2:] = ret[val2:] - ret[:-val2]
    return ret[val2 - 1:] / val2

val1=np.arange(1,26)
val1=np.array(val1)
coso= mediaMovil(val1, val2=10)
print(coso)
````

````python
import numpy as np
import matplotlib.pyplot as plt
"""La señal no tiene cambios, sin embargo al devolverte la media que basicamente es un filtro pasa bajos
para quitar el ruido existente y mejorar la visualizacion de picos"""
def mediaMovil(val1, val2):
    ret = np.cumsum(val1, dtype=float)
    ret[val2:] = ret[val2:] - ret[:-val2]
    return ret[val2 - 1:] / val2
file = open("files/pleth.txt", "r")
pleth = np.array(file.read().split(), dtype = np.float64)
media= mediaMovil(pleth, val2=50)
print(media)
plt.plot(pleth[:1000],scalex=200)
plt.plot(media[:1000],scalex=200)
plt.show()
````
